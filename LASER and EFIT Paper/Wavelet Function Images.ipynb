```python
import pywt
from pywt._extensions._pywt import (DiscreteContinuousWavelet, ContinuousWavelet,
                                Wavelet, _check_dtype)
```

```python
WvltFam = pywt.families()
Wvlts = []
for Fam in WvltFam:
    temp = pywt.wavelist(Fam)
    for wvlt in temp:
        Wvlts.append(wvlt)
Wvlts
```

```python
Waves = ['db3','sym3','coif3','bior2.2','mexh','morl','cgau5','gaus3']
```

```python
i=0
j=0
fig,axs = plt.subplots(4,2,figsize=(4,6), dpi=900)
plt.subplots_adjust(hspace=0.35)

for ax, wave in zip(axs.flatten(), Waves):
    try:
        [psi,x]=pywt.ContinuousWavelet(wave).wavefun(10)
        ax.plot(x,np.real(psi), linewidth=0.5)
    except:
        psi=pywt.Wavelet(wave).wavefun(10)
        ax.plot(np.real(psi[0]), linewidth=0.5)
    ax.set_title(wave, fontsize=7)
    ax.tick_params(axis='both',which='major',labelsize=5)
    ax.tick_params(labelleft=False, left=False)
    i+=1
    if i==4:
        j+=1
        i=0
plt.show()
```
